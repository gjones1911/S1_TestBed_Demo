# FFT

## data gen

```python
## data generation
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Simulation Parameters
sampling_rate = 1000  # Hz (samples per second)
duration = 1  # seconds
base_freq = 59  # Hz

# Harmonic Frequencies
harmonics = [2, 3, 5, 7]  # Multipliers for harmonics
frequencies = [base_freq * h for h in harmonics]

# Time Vector
t = np.linspace(0, duration, sampling_rate * duration, endpoint=False)

# Generate Signal with Harmonics
signal = np.sin(2 * np.pi * base_freq * t)  # Base frequency
for f in frequencies:
    signal += np.sin(2 * np.pi * f * t)

# Create DataFrame
df = pd.DataFrame({'timestamp': t, 'signal': signal})

# Display Data
```

```python
df.head
```

```python
import matplotlib.pyplot as plt

# Plot the Signal against Time
plt.figure(figsize=(10, 5))
plt.plot(df['timestamp'], df['signal'])
plt.xlabel('Time (s)')
plt.ylabel('Signal')
plt.title('Signal vs Time')
plt.grid()
plt.show()
```

```python
# Compute FFT
fft_result = np.fft.fft(signal)  # FFT output (complex values)
time_diff = df['timestamp'].iloc[1] - df['timestamp'].iloc[0]
frequencies = np.fft.fftfreq(len(signal), d=time_diff)  # Frequency bins

# Take the Magnitude Spectrum (absolute value of FFT result)
magnitude = np.abs(fft_result)

# Plot the FFT Spectrum
plt.figure(figsize=(10, 5))
plt.plot(frequencies[:len(frequencies)//2], magnitude[:len(magnitude)//2])  # Only positive frequencies
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.title('FFT Spectrum')
plt.grid()
plt.show()
```
